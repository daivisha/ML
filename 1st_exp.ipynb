```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/DOC-20250403-WA0015_')
print(data)
concepts = np.array(data)[:,:-1]
print(concepts)
target = np.array(data)[:,-1]
print(target)
def train(con, tar):
    # Initialize specific_h before the loop to a default value
    specific_h = None  # Or any other suitable default value
    for i, val in enumerate(tar):
        if val == 'yes':
            specific_h = con[i].copy()
            break
    return specific_h
print(train(concepts, target))
```
